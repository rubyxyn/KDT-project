```python
import pandas as pd
```

### 전처리

```python
news1 = pd.read_excel('./NewsResult_20230101-20230115.xlsx')
news2 = pd.read_excel('./NewsResult_20230116-20230330.xlsx')
news3 = pd.read_excel('./NewsResult_20230331-20230608.xlsx')
news4 = pd.read_excel('./NewsResult_20230609-20230817.xlsx')
news5 = pd.read_excel('./NewsResult_20230818-20231031.xlsx')
```

```python
# 파일 병합
result = pd.concat([news1, news2, news3, news4, news5], axis=0)
result
```

```python
result.to_csv('news230101-231031.csv', encoding='utf-8', index=False)
```

```python
# 파일 불러오기
newsDF = pd.read_csv('./news230101-231031.csv', encoding='utf-8')
```

```python
newsDF.head(1)
```

### 경상일보, 국제신문 제외

```python
filt1 = newsDF['언론사'] != '경상일보'
filt2 = newsDF['언론사'] != '국제신문'

newsDF = newsDF[filt1 & filt2]
newsDF
# newsDF['언론사'].value_counts()
```

```python
newsDF.to_csv('news230101-231031.csv', encoding='utf-8', index=False)
```
